```python
data = pd.read_csv('travel.csv')
```

```python
data
```

```python
# A. What is the distribution of customer ages, and how does age relate to product purchase (ProdTaken)?

data['Age'].plot(kind='kde')   # This is the plot distribution of ages in point column
```

```python
# plot distribution of Age in points column using histogram
data['Age'].plot(kind='hist', edgecolor='black')
```

```python
pd.crosstab(data.Age, data.ProdTaken)  # Here is the Exploratory data analysis of different age groups and their Product Purchase Frequency 
```

```python
# B. How does the type of contact(type of contact)influence the likelihood of purchasing a travel product? 

pd.crosstab(data.TypeofContact, data.ProdTaken)   #(Here, it is showing that self enquiry based customers have taken more products as compared to the invited by the company)
```

```python
# C. Is there a correlation between the city tier(citytier)and monthly income (monthly income) of customers?
data
```

```python
print(data['CityTier'].corr(data['MonthlyIncome']))  # (Here, there is no correlation is showing between CityTier and Monthly Income of the consumers)
```

```python
# D. What is the average duration of a pitch (duration of pitch)for customers who purchased a product versus those who did not?
mean_of_DurationOfPitch = data['DurationOfPitch'].mean()

mean_of_DurationOfPitch
```

```python
# E. How do occupation types(occupation)distribute among customers,and is there an occupation type that is more likely to purchase a travel product? 

pd.crosstab(data.Occupation, data.ProdTaken)  # (Salaried occupation type is more likely to purchase a travel product)
```

```python
# F. Are there any gender-based preferences for the type of product pitched (ProductPitched)? 
prefer_product = pd.crosstab(data['Gender'], data['ProductPitched'])
prefer_product.plot(kind='bar')
```

```python
# G. What is the relationship between the number of trips(number of trips)taken by a customer and their likelihood to purchase a new travel product?
data
```

```python
pd.crosstab(data.NumberOfTrips, data.PreferredPropertyStar)
```

```python
# H. Does having a passport(passport)correlate with a higher number of trips taken or a higher likelihood of purchasing a travel product? 
print(data['Passport'].corr(data['NumberOfTrips']))   # (No, there is no correlation is seeming between Passport and the Number of Trips taken or higher likelyhood of purchasing a travel product) 
```

```python
# I. How satisfied are customers with the pitch (pitch satisfaction score),and does this satisfaction influence product purchase? 

pd.crosstab(data.PitchSatisfactionScore, data.NumberOfFollowups)  # As we can see the PitchSatisfactionScore of 3, 4 and 5 have relatively greater number of followups and this shows that PitchSatisfactionScore very much influence product purchase.
```

```python
# J. Among customers with children (number of children visiting), how does the number of children impact travel product purchases? 
 
children_prodtaken = data.groupby(['NumberOfChildrenVisiting'])['ProdTaken'].sum()
children_prodtaken
```

```python
# K. Gender preferences for product types: how does gender influence the preference for different types of travel products pitched (product pitched)?
pd.crosstab(data.Gender, data.ProductPitched)
```

```python
# L. Impact of marital status on travel preferences: does marital status affect the choice of travel product, and if so, how do different marital statuses correlate with product purchases (prod taken)?

pd.crosstab(data.MaritalStatus, data.ProdTaken) # As we can see from the below table that Product Purchased is in the order of Divorced < Unmarried < Single < Married
```

```python
# M. Relationship between number of children and travel product interest: how does the number of children visiting (number of children visiting) impact the likelihood of purchasing a travel product?
 
pd.crosstab(data.NumberOfChildrenVisiting, data.PreferredPropertyStar)
```

```python
# N. Influence of owning a car on travel decisions: is there a correlation between owning a car (own car) and the number of trips taken (number of trips) or the type of travel product purchased?  

data
```

```python
print(data['OwnCar'].corr(data['NumberOfTrips'])) # It's really a negative correlation between the car owners and the no. of trips they made to purchase a product 
 
```

```python
# O. Income level and product choice: how does the monthly income (monthly income) of customers influence their choice of travel products, and is there a preferred product for different income levels? 
pd.crosstab(data.MonthlyIncome, data.ProductPitched)

```
